```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
Food = pd.read_csv("FoodPrice.csv")
print(Food)
```

```python
NairaPrice = Food["price"].iloc[1:6].sort_values()
UsdPrice = Food["usdprice"].iloc[1:6].sort_values()
Commodity = Food['commodity'].iloc[1:6].sort_values()
```

```python
plt.figure(figsize = (20,5))
plt.subplot(131)
plt.plot(Commodity, NairaPrice,"r--")
plt.xlabel("Commodities")
plt.ylabel("Price in Naira")
plt.title("A Graph of Food Price in Nigeria(NGN)")
plt.legend("Price in NGN")
plt.grid()
plt.subplot(132)
plt.plot(Commodity, UsdPrice,"co")
plt.xlabel("Commodities")
plt.ylabel("Price in Usd")
plt.title("A Graph of Food Price in Nigeria(Usd)")
plt.legend("Price in USD")
plt.grid()
plt.show()
```

```python
plt.pie(NairaPrice, labels = Commodity)
plt.title("A Graph of Food Price in Nigeria(NGN)")
plt.show()

```

```python
#Changing the start angle
plt.pie(NairaPrice, labels = Commodity, startangle = 90)
plt.title("A Graph of Food Price in Nigeria(NGN)")
plt.show()

```

```python
#Exploding out a particular series
#Changing the colours of each slice
Series = [0.5,0,0,0,0]
Colours = [ "red", "orange", "yellow", "green","blue"]
```

```python
plt.pie(NairaPrice, labels = Commodity, startangle = 90, explode = Series, colors = Colours)
plt.title("A Graph of Food Price in Nigeria(NGN)")
plt.legend(title = "Food Commodities")
plt.show()
```
