```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,2,100)

plt.plot(x,x,label="inner")
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("A graph of linear,quadratic and cubic")
plt.legend()
plt.show()
```
