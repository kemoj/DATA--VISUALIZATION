```python
fdp=pd.read_csv("foodprices.csv")
print(fdp)
```

```python
fdp.head()
```

```python
nairap=fdp["price"] 
print(nairap)
```

```python
nairaprice=fdp["price"].iloc[1:6] 
print(nairaprice)
```

```python
usdp=fdp["usdprice"]
print(usdp)
```

```python
usdprice=fdp["usdprice"].iloc[1:6]
print(usdprice)
```

```python

#we are trying to visualise for the 1 to 5 roww
commodity=fdp["commodity"].iloc[1:6]
print(commodity)
```

```python
plt.figure()
plt.subplot()
plt.plot(commodity,nairaprice)
plt.xlabel("commodity")
plt.ylabel("price in naira")
plt.title("A graph of food price in Nign")
plt.show()

plt.subplot()
plt.plot(commodity,usdprice)
plt.xlabel("commodity")
plt.ylabel("price in usd")
plt.title("A graph of food price in usd")
plt.show()
```
