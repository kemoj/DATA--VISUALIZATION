```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hiv= pd.read_excel("HIV_Epidemiology_Children_Adolescents_2021.xlsx", "Data")
hiv
```

```python
Hiv=hiv.rename(columns=hiv.iloc[0]).loc[1:]
Hiv.head(10)
```

```python
Hiv['Indicator'].unique()
```

```python
indicator1=Hiv[Hiv["Indicator"]=='Estimated incidence rate (new HIV infection per 1,000 uninfected population)']
indicator1
```

```python
indicator1.to_csv("indicator1.csv",index=None)
```

```python
Ind1=indicator1[(indicator1["Country/Region"]== "Global") & (indicator1["Sex"]=="Both") & (indicator1["Age"]=="Age 0-14")]
Ind1
```

```python
x= Ind1["Year"]
y=Ind1["Value"]
plt.figure(figsize =(10,6))
plt.plot(x,y)
plt.xlabel("Year")
plt.ylabel("Value")
plt.title('The Graph of Hiv in Children and adolescence Age 0-14 in both Sex Across Year 2000 - 2020')
plt.show()
```

```python
x= Ind1["Year"]
y=Ind1["Value"]
plt.figure(figsize =(10,6))
plt.plot(x,y)
plt.xlabel("Year")
plt.ylabel("Value")
plt.title('The Graph of Hiv in Children and adolescence Age 0-14 in both Sex Across Year 2000 - 2020')
plt.gca().invert_yaxis()
```
