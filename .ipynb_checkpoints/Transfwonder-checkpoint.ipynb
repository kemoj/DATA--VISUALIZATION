```python
from spellchecker import SpellChecker
```

```python
spell = SpellChecker()
mispell = spell.unknown(["let", "us","wlak", "on", "the", "groun"])
for word in mispell:
    print (spell.correction(word))
    print (spell.candidates(word))
```
