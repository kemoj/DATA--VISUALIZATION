```python
local=adap["Local Authority"].iloc[0:11]
value=adap["Value of payments € 2014"].iloc[0:11]
plt.figure(figsize=(20,15))
sns.barplot(x=local, y=value,data=adap)
plt.grid()
sns.set_style("whitegrid")
sns.set_context("poster")
sns.color_palette(palette=None, n_colors=None, desat=None)
plt.show()
```

```python
local=adap["Local Authority"].iloc[0:11]
value=adap["Value of payments € 2014"].iloc[0:11]
plt.figure(figsize=(20,15))
sns.barplot(x=local, y=value,alpha=.9,data=adap)
plt.grid()
sns.set_style("whitegrid")
sns.set_context("paper")
sns.palplot(sns.color_palette("Blues"))
plt.show()
```

```python
value=adap["Value of payments € 2015"].iloc[0:11]
plt.figure(figsize=(10,5))
sns.histplot(x=value)
plt.show()
```
